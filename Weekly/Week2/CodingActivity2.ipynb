```python
#Import packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Import data files [time(s),neutron star mass(gram), correct PNS radius (cm)]
s12Data = np.loadtxt('s12_o0_2D_gw2.dat', usecols = (0,20,29)) #Column 12 from row 0-7123 weird value?
s40Data = np.loadtxt('s40_o0_2D_gw2.dat', usecols = (0,20,29)) #Column 12 from row 0-7123 weird value?
```

```python
# print(s12Data[:10])
# print(s40Data[:10])
```

```python
#Data manipulation
#Adjust Time
s12Data[:,0] = s12Data[:,0] - 0.184469701 
s40Data[:,0] = s40Data[:,0] - 0.366373606 
```

```python
#Plotting the 'Neutron Star Mass' vs time from the data
fig, ax1 = plt.subplots()
fig.set_size_inches(15,15)

ax1.plot(s12Data[:,0], s12Data[:,1] * (5e-34)) #Convert grams to solar masses

ax1.set_ylabel('Neutron Star Mass (solar masses)')
ax1.set_xlabel('Time (s)')
ax1.set_title('s12Data Neutron Star Mass over Time')
```

```python
#Plotting the 'Neutron Star Mass' vs time from the data
fig, ax1 = plt.subplots()
fig.set_size_inches(15,15)

ax1.plot(s40Data[:,0], s40Data[:,1] * (5e-34)) #Convert grams to solar masses

ax1.set_ylabel('Neutron Star Mass (solar masses)')
ax1.set_xlabel('Time (s)')
ax1.set_title('s40Data Neutron Star Mass over Time')
ax1.set_xlim([-0.05,0.35]) #Adjust plot limits for readability sake
```

**Just confirm that the reason mass is increasing is because the neutron stars are 'forming' as the supernova evolves**

```python
#Plotting the 'Correct PNS Radius' vs time from the data
fig, ax1 = plt.subplots()
fig.set_size_inches(15,15)

ax1.plot(s12Data[:,0], s12Data[:,2] * (1.436e-11)) #Converrt cm to solar radii

ax1.set_ylabel('Correct PNS Radius (solar radii)')
ax1.set_xlabel('Time (s)')
ax1.set_title('s12Data Correct PNS Radius over Time')
```

```python
#Plotting the 'Correct PNS Radius' vs time from the data
fig, ax1 = plt.subplots()
fig.set_size_inches(15,15)

ax1.plot(s40Data[:,0], s40Data[:,2] * (1.436e-11)) #Converrt cm to solar radii

ax1.set_ylabel('Correct PNS Radius (solar radii)')
ax1.set_xlabel('Time (s)')
ax1.set_title('s40Data Correct PNS Radius over Time')
ax1.set_xlim([-0.05,0.35]) #Adjust plot limits
```

```python
#Function to find the derivative of a set of values (radius) with respect to time
def derivRadiusWRT(radiusSet,time):
    #Use np.diff to simplify the calculation of differnces
    return(np.diff(radiusSet) / np.diff(time)) #Return an array of the calculated derivatives
```

```python
#Plot the derivatives for 'correct PNS radius' data WRT time, over time
fig, ax1 = plt.subplots()
fig.set_size_inches(15,15)

ax1.plot(s12Data[:-1,0], derivRadiusWRT(s12Data[:,2],s12Data[:,0]) * (1.436e-11)) #Converrt cm to solar radii

ax1.set_ylabel('Correct PNS Radius (solar radii)')
ax1.set_xlabel('Time (s)')
ax1.set_title('s12Data, Derivative of Correct PNS Radius over Time')
```

```python
#Plot the derivatives for 'correct PNS radius' data WRT time, over time
fig, ax1 = plt.subplots()
fig.set_size_inches(15,15)

ax1.plot(s40Data[:-1,0], derivRadiusWRT(s40Data[:,2],s40Data[:,0]) * (1.436e-11)) #Converrt cm to solar radii

ax1.set_ylabel('Correct PNS Radius (solar radii)')
ax1.set_xlabel('Time (s)')
ax1.set_title('s40Data, Derivative of Correct PNS Radius over Time')
ax1.set_xlim([-0.05,0.35]) #Adjust plot limits for readability sake
```
