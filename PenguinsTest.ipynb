```python
import pandas as pds
import matplotlib.pyplot as plt
```

```python
#Import file and 
penguinTest = pds.read_csv('penguins.csv', sep = ',')

penguinTest.head()
```

```python
#General data manipulation
penguinTestMale = penguinTest[penguinTest['sex'] == 'MALE']
penguinTestFemale = penguinTest[penguinTest['sex'] == 'FEMALE']

#Data for plot1
maleFlipLength = penguinTest['flipper_length_mm'][penguinTest['sex'] == 'MALE']
femaleFlipLength = penguinTest['flipper_length_mm'][penguinTest['sex'] == 'FEMALE']

#Data for plot2
maleBillLenIsland = penguinTestMale.iloc[:,1:3]
femaleBillLenIsland = penguinTestFemale.iloc[:,1:3]

#Data for plot3
#Just base array
```

```python
#Plot1 
#print(maleFlipLength, femaleFlipLength)

fig, ax1 = plt.subplots()
fig.set_size_inches(15,15)

ax1.plot(maleFlipLength.sort_values(0, ignore_index = True), label = 'Male', c = 'blue')
ax1.plot(femaleFlipLength.sort_values(0, ignore_index = True), label = 'Female', c = 'green')

ax1.set_ylabel('Flipper Length (mm)')
ax1.set_title('Differences of Flipper Length in Gender')
ax1.xaxis.set_visible(False)
ax1.legend()
```

```python
#Plot2

fig, ax2 = plt.subplots(2)
fig.set_size_inches(15,7)
fig.subplots_adjust(hspace=0.5)

ax2[0].scatter(maleBillLenIsland.iloc[:,1], maleBillLenIsland.iloc[:,0], c = 'blue')
ax2[1].scatter(femaleBillLenIsland.iloc[:,1], femaleBillLenIsland.iloc[:,0], c = 'green')

ax2[0].set_ylabel('Island')
ax2[0].set_xlabel('Bill Length (mm)')
ax2[0].set_title('Differences in Bill Length on different islands (Male)')


ax2[1].set_ylabel('Island')
ax2[1].set_xlabel('Bill Length (mm)')
ax2[1].set_title('Differences in Bill Length on different islands (Female)')
```

```python
#Plot3
fig, ax3 = plt.subplots()
fig.set_size_inches(15,15)

ax3.scatter(penguinTestMale['body_mass_g'], penguinTestMale['bill_depth_mm'], c = 'blue', alpha = 0.6, label = 'Male')
ax3.scatter(penguinTestFemale['body_mass_g'], penguinTestFemale['bill_depth_mm'], c = 'red', alpha = 0.6, label = 'Female')

ax3.set_ylabel('Bill Depth (mm)')
ax3.set_xlabel('Body Mass (g)')
ax3.set_title('Bill Depth compared to Body Mass')
ax3.legend()
```
